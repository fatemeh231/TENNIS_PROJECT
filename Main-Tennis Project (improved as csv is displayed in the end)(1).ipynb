Question #2: what is the average height of the players?

```python
All_players.drop_duplicates('full_name',inplace=True)
All_players.info()
All_players.dropna(inplace=True)
All_players_height_Ave=All_players['height'].mean()
print(f'The Average height of the players is: ' + str(round(All_players_height_Ave,2)) + ' cm')
All_players.to_csv('heights_of_players.csv', index=False)
```
